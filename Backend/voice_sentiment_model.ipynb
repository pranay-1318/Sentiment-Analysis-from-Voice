```python
import speech_recognition as sr
import nltk
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC

nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
# Text Preprocessing
def preprocess_text(text):
    text = text.lower()
    text = re.sub('[^a-zA-Z0-9 ]', '', text)
    words = text.split()
    words = [w for w in words if w not in stopwords.words('english')]
    return " ".join(words)
```

```python
sentences = [preprocess_text(s) for s in final_dataset['sentence']]
```

```python
sentences
```

```python
emotion = final_dataset['emotion']
emotion
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# train_test_split() SHUFFLES the data by default during split
X_train, X_test, y_train, y_test = train_test_split(
    sentences,emotion, test_size=0.25, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
# Vectorizer
vectorizer = TfidfVectorizer(max_features=3000)
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
# Model Training
model = LinearSVC()
model.fit(X_train, y_train)
```

```python
# Model Evaluation (Accuracy)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
# Speech to Text
def speech_to_text(audio_file):
    recognizer = sr.Recognizer()
    with sr.AudioFile(audio_file) as source:
        audio = recognizer.record(source)
        text = recognizer.recognize_google(audio, language="en-US")
    return text
```

```python
speech_to_text("/content/sad.wav")
```

```python
# Final Prediction Function
def predict_emotion(audio_file):
    text = speech_to_text(audio_file)
    processed_text = preprocess_text(text)
    vector = vectorizer.transform([processed_text])
    emotion = model.predict(vector)[0]
    return text, emotion
```

```python
# unseen data
predict_emotion("/content/sad.wav")
```
